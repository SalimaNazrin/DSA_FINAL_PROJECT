## Exploratory Data Analysis

EDA stands for Exploratory Data Analysis. It's an approach to analyzing datasets to summarize their main characteristics, often employing visual methods. The primary goal of EDA is to understand the data and its underlying structure, patterns, and relationships, without making any assumptions about statistical models

```python
disaster_category_df = pd.DataFrame(data.iloc[:,1:4].groupby(['Disaster Subgroup','Disaster Type']).value_counts(), columns=["Frequency"])

disaster_category_df

```

Finding the most frequent disaster type

```python
disaster_category_df.where(disaster_category_df == disaster_category_df.Frequency.max()).dropna()
```

Finding the least occuring disaster type

```python
disaster_category_df.where(disaster_category_df == disaster_category_df.Frequency.min()).dropna()
```
